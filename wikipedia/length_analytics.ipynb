```python
import matplotlib.pyplot as plt

with open('simple_wikipedia_sentences.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

line_lengths = [len(line.strip()) for line in lines]

plt.figure(figsize=(10, 6))
plt.plot(line_lengths, marker='o', linestyle='-', color='b')
plt.title('Line Length Analysis')
plt.xlabel('Line Number')
plt.ylabel('Length of Line')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(line_lengths, bins=1000)
plt.title('Line Length Frequency Histogram')
plt.xlabel('Length of Line')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
